```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Data Cleansing

### Import CSV

```python
df = pd.read_csv("income.csv")
df.head()
```

### Checking data types

```python
df.dtypes
```

### Drop unnecesary column

```python
df = df.drop(['workclass', 'fnlwgt', 'educational-num', 'marital-status', 'relationship', 'capital-gain', 'capital-loss', 'hours-per-week'], axis=1)
df.head()
```

### Renaming column

```python
df = df.rename(columns={"native-country" : "country"})
df.info()
```

### Drop duplicated data

```python
dup = df[df.duplicated()]
dup.shape
```

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)
df.shape
```

### Change the categorical data into numeric data

```python
df = pd.get_dummies(df[['occupation', 'race', 'gender', 'country', 'income']],drop_first=False)
df
```

### Combining translated lable column into one

```python
df['income'] = pd.Series(df[['income_<=50K', 'income_>50K']].values.argmax(axis=1))
df = df.drop(['income_<=50K', 'income_>50K'], axis=1)

df
```

### Slice the data for lable and data for testing

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

y
```

# Modeling

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clas = KNeighborsClassifier(n_neighbors=6, metric="euclidean", p=2)
clas.fit(x_train, y_train)
```

```python
y_pred = clas.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn import metrics
print("Accuracy : ", metrics.accuracy_score(y_test, y_pred))
```
